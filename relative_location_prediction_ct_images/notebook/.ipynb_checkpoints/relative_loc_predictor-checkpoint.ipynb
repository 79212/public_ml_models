```python
import pandas as pd

df = pd.read_csv('../data/slice_localization_data.csv')
df.head()
```

```python
df = df.drop(['patientId'], axis=1)
df_y = df['reference']
df_x = df.drop(['reference'], axis=1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df_x = scaler.fit_transform(df_x)
pca = PCA(0.95)
pca_vectors = pca.fit_transform(scaled_df_x)
for index, var in enumerate(pca.explained_variance_ratio_):
    print("Explained Variance ratio by Principal Component ", (index+1), " : ", var)
```

```python
pca_vectors
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="ticks")
sns.set_palette(palette='Set1')

fig_1 = plt.figure(figsize=(18,6))

sns.set(style="ticks")
sns.set_palette(palette='Set1')

sns.regplot(x=pca_vectors[:,0],y=df_y, label='Principal Component 1',x_bins=10)
sns.regplot(x=pca_vectors[:,1],y=df_y, label='Principal Component 2',x_bins=10)
sns.regplot(x=pca_vectors[:,2],y=df_y, label='Principal Component 3',x_bins=10)

plt.title('Most Important Principal Components vs Reference Value')
plt.xlabel('Principal Component Value')
plt.ylabel('Reference Value')
plt.legend()
plt.show()
```

```python
fig_2 = plt.figure(figsize=(18,6))

sns.regplot(x=pca_vectors[:,209],y=df_y, label='Principal Component 210',x_bins=10)
sns.regplot(x=pca_vectors[:,210],y=df_y, label='Principal Component 211',x_bins=10)
sns.regplot(x=pca_vectors[:,211],y=df_y, label='Principal Component 212',x_bins=10)

plt.title('Least Important Principal Components vs Reference Value')
plt.xlabel('Principal Component Value')
plt.ylabel('Reference Value')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(pca_vectors, df_y)
```

```python
from sklearn.linear_model import ElasticNetCV

regr_en = ElasticNetCV(cv=5, alphas=[0.1, 0.3, 0.5, 0.7, 1.0])
regr_en.fit(train_x, train_y)
r2_en = regr_en.score(X=test_x, y=test_y)
```

```python
print('R2 value : ', r2_en)
```

```python
print('Intercept: ', regr_en.intercept_) 
print('Alpha: ', regr_en.alpha_) 
print('Coefficients: ' , regr_en.coef_) 
```

```python
fig_3 = plt.figure(figsize=(18,6))

sns.regplot(x=test_x[:,0],y=regr_en.predict(test_x), label='Principal Component 1',x_bins=10)
sns.regplot(x=test_x[:,1],y=regr_en.predict(test_x), label='Principal Component 2',x_bins=10)
sns.regplot(x=test_x[:,2],y=regr_en.predict(test_x), label='Principal Component 3',x_bins=10)

plt.title('Most Important Principal Components vs Estimated <Reference> Value')
plt.xlabel('Principal Component Value')
plt.ylabel('Estimated <Reference> Value')
plt.legend()
plt.show()
```

```python
fig_4 = plt.figure(figsize=(18,6))

sns.regplot(x=test_x[:,209],y=regr_en.predict(test_x), label='Principal Component 210',x_bins=10)
sns.regplot(x=test_x[:,210],y=regr_en.predict(test_x), label='Principal Component 211',x_bins=10)
sns.regplot(x=test_x[:,211],y=regr_en.predict(test_x), label='Principal Component 212',x_bins=10)

plt.title('Least Important Principal Components vs Estimated <Reference> Value')
plt.xlabel('Principal Component Value')
plt.ylabel('Estimated <Reference> Value')
plt.legend()
plt.show()
```

```python
residuals = test_y - regr_en.predict(test_x)
```

```python
fig_5 = plt.figure(figsize=(18,6))

sns.scatterplot(data=residuals.values)
plt.title('Residual Plot', fontsize=16)
plt.ylabel('Residual', fontsize=16)
plt.xlabel('Data Instance', fontsize=16)
plt.show()
```

```python
from sklearn.pipeline import Pipeline

pl_train_x, pl_test_x, pl_train_y, pl_test_y = train_test_split(df_x, df_y)

pl_model = Pipeline([('scaler', StandardScaler()),
                     ('pca', PCA(0.95)),
                     ('regr', ElasticNetCV(cv=5, alphas=[0.1, 0.3, 0.5, 0.7, 1.0]))])

pl_model.fit(pl_train_x, pl_train_y)
```

```python
from joblib import dump, load

model_descriptor_name = 'pc_elasticnet_model.joblib'
dump(pl_model, model_descriptor_name)
saved_model = load(model_descriptor_name)
saved_model.predict(pl_test_x[20:21])
```
